## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform

# Uniform distribution
a = 0  # lower bound
b = 10 # upper bound

# loc = a (start point), scale = b-a (range)
random_numbers = uniform.rvs(loc=a, scale=b-a, size=1000)

print(f"First 10 random numbers: {random_numbers[:10]}")
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code herefrom scipy.stats import uniform
import matplotlib.pyplot as plt

# 1. Function to generate uniform random numbers
def generate_uniform(bottom, ceiling, count):
    random_numbers = uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)
    return random_numbers
     
```

```python
# Call function with 2 sets of parameters
data1 = generate_uniform(bottom=10, ceiling=15, count=100)
data2 = generate_uniform(bottom=10, ceiling=60, count=1000)
print(f"First 10 random numbers (10-15): {data1[:10]}")
print(f"First 10 random numbers (10-60): {data2[:10]}") 
```

```python
# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# First histogram
axes[0].hist(data1, bins=10, color='skyblue', edgecolor='black')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Count')
axes[0].set_title('Uniform Distribution (10, 15, 100)')
axes[0].grid(axis='y', alpha=0.3)

# Second histogram
axes[1].hist(data2, bins=10, color='lightcoral', edgecolor='black')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Count')
axes[1].set_title('Uniform Distribution (10, 60, 1000)')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

How are the two distributions different?

100 samples distributed over small range. 

1000 samples (10x more data)

More samples (1000 vs 100) â†’ higher bar counts

Summary: Same distribution type (uniform), but different ranges and sample sizes

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# 1. Function to generate normal random numbers
def generate_normal(mean, std_dev, count):
    random_numbers = norm.rvs(loc=mean, scale=std_dev, size=count)
    return random_numbers
```

```python
# 2. Generate 1000 numbers with mean=10, std=1
data1 = generate_normal(mean=10, std_dev=1, count=1000)
```

```python
# 3. Generate 1000 numbers with mean=10, std=50
data2 = generate_normal(mean=10, std_dev=50, count=1000)
```

```python
# 4. Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# First histogram (mean=10, std=1)
axes[0].hist(data1, bins=30, color='steelblue', edgecolor='black')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Count')
axes[0].set_title('Normal Distribution (mean=10, std=1)')
axes[0].grid(axis='y', alpha=0.3)

# Second histogram (mean=10, std=50)
axes[1].hist(data2, bins=30, color='steelblue', edgecolor='black')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Count')
axes[1].set_title('Normal Distribution (mean=10, std=50)')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# 1. Function to generate exponential distribution
def generate_exponential(mean, size):
    # scale parameter = mean for exponential distribution
    numbers = np.random.exponential(scale=mean, size=size)
    return numbers

# 2. Generate two sequences
data1 = generate_exponential(mean=10, size=10)
data2 = generate_exponential(mean=10, size=100)

# 3. Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# First histogram (size=10)
axes[0].hist(data1, bins=100, color='skyblue', edgecolor='black')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Count')
axes[0].set_title('Exponential Distribution (mean=10, size=10)')
axes[0].grid(axis='y', alpha=0.3)

# Second histogram (size=100)
axes[1].hist(data2, bins=100, color='lightcoral', edgecolor='black')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Count')
axes[1].set_title('Exponential Distribution (mean=10, size=100)')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 
Small size â†’ choppy, irregular histogram.
Large size â†’ smooth, clear pattern

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. Î» = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
from scipy.stats import expon

# Given information
mean = 10  # minutes
lambda_rate = 1 / mean  # Î» = 1/10 = 0.1
# We want P(X < 15)
x = 15

# Method 1: Using scipy.stats.expon
# CDF gives P(X <= x)
probability = expon.cdf(x, scale=mean)

print(f"Mean time in bank: {mean} minutes")
print(f"Lambda (Î»): {lambda_rate}")
print(f"Probability that customer spends less than 15 minutes: {probability:.4f}")
print(f"Probability in percentage: {probability * 100:.2f}%")



# Using formula P(X < x) = 1 - e^(-Î»x)
probability_formula = 1 - np.exp(-lambda_rate * x)
print(f"\nUsing formula: {probability_formula:.4f}")
```

What is the probability that the customer will spend more than 15 minutes

```python
# Given information
mean = 10  # minutes
lambda_rate = 1 / mean  # Î» = 1/10 = 0.1
x = 15

# P(X > 15) = 1 - P(X <= 15)
prob_less_than_15 = expon.cdf(x, scale=mean)
prob_more_than_15 = 1 - prob_less_than_15

print(f"Probability that customer spends MORE than 15 minutes: {prob_more_than_15:.4f}")
print(f"Probability in percentage: {prob_more_than_15 * 100:.2f}%")
```
